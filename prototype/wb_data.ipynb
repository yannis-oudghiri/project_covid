```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
```

```python
# List of countries with their ISO codes
COUNTRIES = {
    "USA": "États-Unis",
    "CHN": "Chine",
    "IND": "Inde",
    "BRA": "Brésil",
    "RUS": "Russie",
    "GBR": "Royaume-Uni",
    "FRA": "France",
    "DEU": "Allemagne",
    "JPN": "Japon",
    "KOR": "Corée du Sud",
}

# Dictionary of indicators 
INDICATORS = {
    "NY.GDP.MKTP.CD": "GDP",
    "SL.UEM.TOTL.ZS": "Unemployment",
    "SH.DYN.MORT": "Mortality",
    "FP.CPI.TOTL.ZG": "InflationRate",
    "ST.INT.ARVL": "International Arrivals",
    "ST.INT.XPND.CD": "TourismExpenditure",
}
```

```python
# Set the date range
data_date = (datetime.datetime(2018, 1, 1), datetime.datetime(2022, 12, 31))

# Retrieve the data
data = wbdata.get_dataframe(INDICATORS, country=list(COUNTRIES.keys()), data_date=data_date)
data
```

```python
# missing values
data.fillna(0, inplace=True)
```

```python
# Group data by country and calculate the mean
data_grouped = data.groupby('country').mean()
data_grouped.reset_index(inplace=True)

# Save the grouped data to a CSV file
data_grouped.to_csv("world_bank_covid_economy_data.csv")
```

```python
# Define a function for plotting 

def plot_data(data, indicator, title):
    plt.figure(figsize=(12, 6))  
    sns.set_style("whitegrid")  

    data_stack = data.stack().reset_index()
    data_stack_gdp = data_stack[data_stack['level_2'] == indicator]
    data_stack_gdp[0] = data_stack_gdp[0].astype(float)

    sns.lineplot(data=data_stack_gdp, x='date', y=0, hue='country', lw=2)  
    plt.title(title, fontsize=16)  
    plt.xlabel("Date", fontsize=10)  
    plt.ylabel("Value", fontsize=14)  
    plt.xticks(fontsize=12)  
    plt.yticks(fontsize=12)  
    plt.legend(title="Country", title_fontsize=12, fontsize=10)  

    
    plt.savefig(f"{indicator}_plot.png", dpi=350)

    plt.show()

# Plotting
plot_data(data, 'GDP', "GDP Over Time")
plot_data(data,'Unemployment', "Unemployment Over Time")
plot_data(data,'Mortality',"Mortality over Time")
plot_data(data, 'TourismExpenditure', "Tourism Expenditure Over Time")
plot_data(data,'International Arrivals', "International Arrivals Over Time")
plot_data(data,'InflationRate', "Inflation rate Over Time")
```

```python
# Calculate correlation between economic indicators
correlation_data = data_grouped.corr()

# Create a heatmap to visualize the correlation
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Economic Indicators")
plt.show()
```
